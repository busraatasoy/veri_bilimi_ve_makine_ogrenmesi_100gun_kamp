## **🔍 Pandas Nedir?**

Pandas, farklı veri kaynaklarından **(örneğin CSV, Excel, SQL, JSON)** gelen verileri kolayca okuyup tablo formatında işler.

## ➡ Series Yapısı Nedir?

Series, tek boyutlu veri saklamak için kullanılır.
Her elemanın bir indeksi (index) vardır.

````python
# Pandas kütüphanesini yükleyelim

import pandas as pd
````

````python
# Öğrenci notlarının sözlük formunda olduğunu varsayalım:

grades = {"Atil": 50, "James": 60, "Lars": 30}
````

````python
# Oluşturduğumuz series yapısına bakalım:

pd.Series(grades)
````

````python
# İki farklı liste halinde oluşturalım

names = ["Atil", "James", "Lars"]
grades = [50, 60, 30]
````

````python
# Names listesini kontrol edelim

pd.Series(names)
````

````python
# grades listesini kontrol edelim

pd.Series(grades)
````

````python
# İki listeye aynı anda bakalım

pd.Series(names, grades)
````

````python
# Argüman ile iki listeye bakalım

pd.Series(data=grades, index=names)
````

````python
# Numpy dizileri ile kullanalım

# Kütüphaneyi yükleyelim
import numpy as np

# array(dizi) oluşturalım
numpy_array = np.array([50,40,30,20])
````

````python
# Numpy dizisinden Pandas Series oluşturalım

pd.Series(numpy_array)
````

## ➡ Series ile Aritmetik İşlemler

````python
# Aritmetik olarak bakalım

# 1. yarışma sonucu
contest_result = pd.Series(data = [10,5,100], index = ["Atil", "James", "Lars"])

contest_result
````

````python
# 2. yarışma sonucu

contest_result2 = pd.Series(data = [20,50,10], index = ["Atil", "James", "Lars"])

contest_result2
````

````python
# Kişi bazlı sonuçlara bakalım

contest_result["Atil"]
````

````python
contest_result2["James"]
````

````python
# Final sonucuna bakalım

final_result = contest_result + contest_result2

final_result
````

````python
# Diğer aritmetik işlemlerde yapılır:

contest_result - contest_result2
````

````python
contest_result * contest_result2
````

````python
contest_result / contest_result2
````

````python
# yeni seri oluşturalım

diff_series = pd.Series([20,30,40,50], ["a", "b", "c", "d"])

diff_series
````

````python
# farklı index elemanlarıyla yeni seri2 oluşturalım

diff_series2 = pd.Series([10,5,3,1], ["a", "c", "f", "g"])

diff_series2
````

````python
# Bu serileri toplayalım
# Verileri eşleşmediğinde& bulamadığında NaN değeri verilir.

diff_series + diff_series2
````

## ➡  DataFrame Yapısı Nedir?

DataFrame, Pandas’ın en güçlü yapısıdır. Satır ve sütunlardan oluşan tablo benzeri bir veri yapısıdır.

````python
# Kütüphaneleri yükleyelim

import pandas as pd
import numpy as np
````

````python
# Basit veri yapısı oluşturalım(random)

data = np.random.randn(4,3)

data
````

````python
# Dataframe oluşturalım

df = pd.DataFrame(data)

df
````

````python
# Tip kontrolü yapalım

type(df)
````

````python
# Tip kontrolü yapalım
# İlk eleman için

type(df[0])
````

````python
# Yeni dataframe oluşturalım
# Kolon isimleriyle oluşturalım

new_df = pd.DataFrame(data, index =["Atil", "Zeynep", "Atlas", "Mehmet"], columns=["Salary", "Age", "Seniority"])

new_df
````

````python
# Yaş kolonuna erişelim

new_df["Age"]
````

````python
# Maaş kolonuna erişelim

new_df["Salary"]
````

````python
# Birden fazla kolona erişelim

new_df[["Salary", "Seniority"]]

````

````python
# Tek bir elemana erişme

#loc → Etiket (isim) bazlı erişim
#loc, etiketleri (index adlarını) kullanarak erişim sağlar.

new_df.loc["Atil"]
````

````python
# Tek bir elemana erişme

# iloc → Konum (index numarası) bazlı erişim
# iloc, satır sırasına (index numarasına) göre erişim sağlar.
# Yani iloc[0] ilk satırı döndürür.

new_df.iloc[0]
````

````python
# Hücre Bazında Erişim
# loc ile:
# Bazen sadece belirli bir hücreye (örneğin Atil’in yaşı) erişmek isteyebilirsin.
# Bunu şu şekilde yapabiliriz: (satır etiketi, sütun adı)

new_df.loc["Atil", "Age"]
````

````python
# Hücre Bazında Erişim
# iloc ile:
# iloc ile (satır index numarası, sütun index numarası)

new_df.iloc[0,1]
````

````python
# Hücre Bazında Erişim
# Bütün satırlara erişmek için( Age)
# iloc ile (satır index numarası, sütun index numarası)

new_df.iloc[:,2]
````

## ➡  Data Manipülasyonu

````python
# Veri setine yeni kolon ekleyelim

new_df["Extra"] = 10

new_df
````

````python
# Eklediğimiz "Extra" sütununu silelim
# axis = 1 sütunu belirtir.
# inplace = True değişikliği verisetinde kalıcı olarak yapar.

new_df.drop("Extra", axis = 1, inplace = True)

new_df
````

````python
# Sütuna "Mehmet" değişkeni ekleyelim

new_df["Mehmet"] = 10

new_df
````

````python
# Mehmet isimli veriyi silelim
# axis belirtilmezse default değer 0 olarak satır(row) belirtir.

new_df.drop("Mehmet")
````

````python
# Atama olmadığı ve inplace= True yapılmadığı için değişiklik kalıcı olmadı

new_df
````

````python
# Mehmet isimli sütunu silelim

new_df.drop("Mehmet", axis = 1, inplace = True)

new_df
````

````python
# Atlas satırının değerlerini alalım

new_df.loc["Atlas"]
````

````python
# Atlas satırının Salary değerini alalım

# new_df.loc["Atlas"]["Salary"] veya

new_df.loc["Atlas", "Salary"]
````

````python
# Atlas'ın Salary değerini değiştirelim

new_df.loc["Atlas", "Salary"] = 100

new_df
````

````python
# Filtreme işlemleri yapalım

boolen_frame = new_df > 0

new_df[boolen_frame]
````

````python
# Aynı filtreleme eşitleme olmadanda doğrudan yapılabilir

new_df[new_df > 0]
````

````python
# Salary için filtreleme yapalım

new_df[new_df["Salary"] > 1 ]
````

````python
# Age için filtreleme yapalım

new_df[new_df["Age"] > 0 ]
````

## ➡  DataFrame Index Mantığı

````python
# Index atayalım

reset_frame = new_df.reset_index()

reset_frame
````

````python
# loc ile indexe erişme

reset_frame.loc[0]
````

````python
# Yeni index ekleyelim

new_indices = ["Ati", "Zey", "Atl", "Meh"]

new_df["NewIndex"] = new_indices

new_df
````

````python
# Fonksiyon ile index oluşturalım

new_df.set_index("NewIndex", inplace = True)

new_df
````

````python
# Yeni indexe göre değeri bulalım

new_df.loc["Ati"]
````

````python
# multi index

#İlk indexi oluşturalım

first_index = ["Simpson", "Simpson", "Simpson", "South Park", "South Park", "South Park"]
````

````python
# İkinci indexi oluşturalım

inner_index = ["Homer", "Bart", "Marge", "Cartman", "Kenny", "Kyle"]
````

````python
# İki indexi birleştirelim

zipped_index = list(zip(first_index, inner_index))

zipped_index
````

````python
# MultiIndex yapısını oluşturalım

zipped_index = pd.MultiIndex.from_tuples(zipped_index)

zipped_index
````

````python
# Örnek data oluşturalım

sample_values = np.ones((6,2))
````

````python
# Dataframe oluşturalım

big_df = pd.DataFrame(sample_values, index = zipped_index, columns = ["Age", "Salary"])

big_df
````

````python
# Age değerlerine ulaşalım

big_df["Age"]
````

````python
# Homer bilgilerine ulaşalım

big_df.loc["Simpson"].loc["Homer"]
````
