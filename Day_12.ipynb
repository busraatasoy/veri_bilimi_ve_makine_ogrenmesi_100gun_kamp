```python
#9) Aşağıdaki kodun çıktısı ne olacaktır?
```

```python
class Ogrenci():

    def __init__(self,isim,sinavNotu):
        self.isim = isim
        self.__sinavNotu = sinavNotu      #encapsulation

    def notuGoster(self):
        print(f"{self.isim} sınav notu: {self.__sinavNotu}")
```

```python
ogrenci = Ogrenci("Mehmet",85)
```

```python
ogrenci.__sinavNotu = 75        # encapsulation sebebiyle bu ekrana gelmez.
```

```python
#ogrenci.notuGoster()
```

```python
# cevap
# 85 sınav notu gösterecek.

ogrenci.notuGoster()
```
