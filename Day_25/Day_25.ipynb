
**Broadcasting Her Boyuta Uygulanabilir mi?**

**Cevap:** Hayır, broadcasting her boyuta otomatik olarak uygulanamaz.

**Broadcasting Kuralları:**

- NumPy’de iki dizinin toplama, çıkarma veya diğer element-wise işlemlerinde **broadcasting yapılabilmesi için**:

- Dizilerin son boyutlarından başlayarak boyut sayıları **eşit** olmalı veya **biri 1** olmalı.

- Boyutlar yukarıdan aşağıya karşılaştırılır ve eğer herhangi bir boyutta bu koşul sağlanmazsa `ValueError` alınır.

## ➡ **Matrix Çarpımları**

**Matris çarpımı(matrix multiplication)**, iki matrisin elemanlarının aynı konumdaki değerlerle değil, **satır ve sütun kombinasyonlarıyla çarpılıp toplanmasıdır.**

Bu işlem element-wise **toplama/çarpma** işlemlerinden **farklıdır.**

Matrix işlemleri özellikle görüntü işleme alanında çok fazla kullanılmaktadır.

**1. Element-wise (Eleman Bazlı) Çarpma**

Aynı boyutlardaki iki matrisin elemanları birebir çarpılır:

````python
first_arr
````

````python
second_arr
````

````python
# first_arr ve second_arr ile çarpma işlemi yapalım

first_arr * second_arr
````

Satır ve sütun değerleri karşılıklı olarak çarpılır.

**2. Gerçek Matris Çarpımı (Dot Product)**

Matematikteki klasik matris çarpımıdır.

Kural: **Birinci matrisin sütun sayısı, ikinci matrisin satır sayısına eşit** olmalıdır.




```
print(np.dot(A, B))
# veya A @ B şeklinde de yazılabilir

```



````python
# Yeni matrix tanımlayalım

first_matrix = np.array([[10,20,30]])

first_matrix
````

````python
# Matrix şeklini kontrol edelim

first_matrix.shape
````

````python
# İkinci matrix tanımlayalım

second_matrix = np.array([[2,3],[2,3],[2,3]])

second_matrix
````

````python
# Matrix şeklini kontrol edelim

second_matrix.shape
````

````python
# Element bazlı çarpımda hata verir
# Satır-sütun uyuşmazlığı

# first_matrix * second_matrix
````

````python
# Dot product ile çarpım işlemi

result_matrix = first_matrix.dot(second_matrix)

result_matrix
````

````python
#Şeklini kontrol edelim

result_matrix.shape
````

````python
# dot yerine @ ile kullanımı ile çarpımı

result_matrix2 = first_matrix @ second_matrix

result_matrix2
````

````python
# Z-Score Hesaplama
# Outlier değerler var.

data = np.array([10,12,13,15,18,25,100,105])
data
````

````python
# Ortalama hesaplama

mean = np.mean(data)

mean
````

````python
# Veri toplamı

data.sum()
````

````python
# Manuel ortalama(veri toplamı/veri sayısı)

data.sum() / len(data)
````

````python
# standart sapma hesaplama

std = np.std(data)

std
````

````python
# Z Score = (Veri- Ortalama) / Standart sapma

z_score = (data - mean) / std

z_score
````

````python
# z score 1'den büyük olanları bulalım
# Boolen formunda gelir

z_score > 1
````

````python
# z score 1'den büyük olanlar değerler hangileri?

z_score[z_score > 1]
````

````python
# z score 1'den büyük olanlar veriler hangileri?
# Outlier değerlerdir.

data[z_score > 1]
````
