## 1. Inheritance (Kalıtım)

**Tanım:**

- Bir sınıfın başka bir sınıftan özellik ve metot miras almasıdır.

**Amaç:**

- Kod tekrarını azaltmak ve ortak işlevleri paylaşmak.

**Normal bir sınıf tanımı yapalım:**

````python
# Müzisyen sınıfı tanımlayalım:

class Musician:
    def __init__(self,name):           #initializer
      self.name = name            #property
      print("musician class")

    #method1
    def test1(self):
        print("test1")

    #method2
    def test2(self):
        print("test2")
````

````python
# busra objesini değer atayarak tanımlayalım:

busra = Musician("Busra")
````

````python
# name çalıştıralım:

busra.name
````

````python
# test1 çalıştıralım:

busra.test1()
````

````python
# test2 çalıştıralım:

busra.test2()
````

**Inheritance(Kalıtım) Kullanarak Önceki Sınıfın Özelliklerini İçeren Yeni bir Sınıf Oluşturalım:**

````python
# Sınıfın aynı işini yapan farklı sınıf tanımlayalım:

# self yerine üstte tanımladığımız Musician sınıfını parantez içinde() yazarız.
class MusicianPlus(Musician):

    def __init__(self):               #initializer
      Musician.__init__(self)         # Musician initini kendi içinde çalıştıralım.
      print("musician plus")
````

````python
# MusicianPlus için değişken tanımlayıp çalıştıralım:
# Kod çalıştırığıldığında hata alırız.

# mustafa = MusicianPlus()
````

**`TypeError:` hatası aldık.** **Bu hata ne anlama geliyor?**

Bu hata, `MusicianPlus` sınıfından bir nesne oluşturulurken` __init__` fonksiyonuna gerekli olan bir argümanın **(yani name)** verilmemesinden kaynaklanır.

Musician sınıfının __init__ metodu şöyle tanımlanmıştı:



```
def __init__(self, name):

```

Bu, `MusicianPlus` sınıfından bir nesne oluştururken mutlaka bir `name` değeri verilmesi gerektiği anlamına gelir.

Ancak biz bu satırı yazdık:



```
mustafa = MusicianPlus()

```

Burada () içinde hiçbir değer verilmedi. Yani Python, name için bir şey bulamıyor ve şöyle diyor:

**“Sen bana name göndermedin ama ben onu bekliyordum!”**



````python
# Kodu derleyerek düzeltelim:

# self yerine üstte tanımladığımız Musician sınıfını parantez içinde() yazarız.
class MusicianPlus(Musician):

    def __init__(self, name):               #initializer
      Musician.__init__(self, name)         # Musician initini kendi içinde çalıştıralım.
      print("musician plus")
````

````python
# MusicianPlus için değişken tanımlayıp name parametre değerini girelim:

mustafa = MusicianPlus("Mustafa")
````

Evet ! Gördüğümüz gibi kodda hem Musician sınıfında print ettiğimiz **"musician class"** hemde MusicianPlus sınıfınfa print ettiğimiz **"musician plus"** yazısı çalıştı. Böylece **kalıtım ile önceki sınıftan değeri çalıştırmış** ve üzerine de **kendi yazdığımız yeni yazıyı** çalıştırmış olduk.

Musician sınıfında tanımlanan `test1` ve `test2` methodlarını da aynı şekilde çalıştırabiliriz.

````python
# Musician sınıfında yazdığımız test mehodlarını çalıştıralım:

mustafa.test1()
````

````python
mustafa.test2()
````

````python
# name değişken atamasını değiştirelim:

mustafa.name = "Mustafa 2"
````

````python
# Güncellenen ismi yazdıralım:

mustafa.name
````

````python
# Tipini kontrol edelim:
# MusicianPlus sınıfında olduğunu anlarız.


type(mustafa)
````

````python
# MusicianPlus içinde test3 methodu oluşturalım:


# self yerine üstte tanımladığımız Musician sınıfını parantez içinde() yazarız.
class MusicianPlus(Musician):

    def __init__(self, name):               #initializer
      Musician.__init__(self, name)         # Musician initini kendi içinde çalıştıralım.
      print("musician plus")

    def test3(self):                        #test3 methodu tanımladık
      print("test3")

````

````python
# MusicianPlus için değişken tanımlayıp name parametre değerini tekrar girelim:

mustafa = MusicianPlus("Mustafa")
````

````python
# Methodu kontrol edelim:

mustafa.test3()
````

Burada `test3`metodu yalnızca **MusicianPlus** sınıfı içinde tanımlıdır. Bu nedenle, **Musician** sınıfının içeriğinde `test3` bulunmaz.

**Bu durum bize şu avantajı sağlar:**

- **MusicianPlus** sınıfından oluşturulan `mustafa nesnesi`, hem **Musician** sınıfından miras aldığı` name`, `test1`, `test2` gibi özelliklere hem de kendi özel metodu olan `test3`'e sahiptir.

- Buna karşın, sadece **Musician** sınıfından oluşturulan bir `busra` nesnesi yalnızca `name`, `test1`, ve `test2` özelliklerine sahiptir. `test3` fonksiyonu bu nesnede yoktur, çünkü sınıfı içinde tanımlı değildir.

Bu da bize kalıtım yoluyla ortak özellikleri yeniden yazmadan kullanma imkânı verirken, aynı zamanda her alt sınıfa özgü özel işlevler de tanımlayabilme esnekliği sağlar.



**Ekstra Bilgi:**

Bu yapı, nesne yönelimli programlamada (OOP) şu avantajları sunar:

1. Kod tekrarını azaltır:

- Ortak fonksiyonları üst sınıfa yazarsın, alt sınıflar otomatik miras alır.

2. Alt sınıflar özelleştirilebilir:

- Her alt sınıf kendi işlevlerini (test3 gibi) ekleyebilir.

3. Daha esnek ve ölçeklenebilir yapılar kurulur:

- Yeni özellik eklemek istendiğinde üst sınıfa dokunmadan alt sınıflar geliştirilebilir.

Peki, ben **Musician** sınıfı içindeki `test1` metodunu **MusicianPlus** içinde değiştirmek istiyorum, yapabilir miyim ?

Evet. Yapabilirim. Buna da `overrade` deniyor.

**Override Nedir?**

Override (üstüne yazmak), bir alt sınıfın, üst sınıf (parent class) içinde tanımlanmış bir metodu kendine göre yeniden tanımlamasıdır.

Bu şekilde, alt sınıf aynı isimde bir metod tanımladığında, üst sınıftaki metod "görünmez" olur ve alt sınıftaki çalışır.

````python
# MusicianPlus içinde test1 methodunu değiştirelim:


class MusicianPlus(Musician):

    def __init__(self, name):               #initializer
      Musician.__init__(self, name)         # Musician initini kendi içinde çalıştıralım.
      print("musician plus")

    def test3(self):                        #test3 methodu tanımladık
      print("test3")

    #override
    def test1(self):                        #test1 methodunu yeniden tanımladık
      print("test1 methodu overrade kullararak değiştirildi")
````

````python
# busra objesini değer atayarak tanımlayalım:

busra = Musician("Busra")
````

````python
# Musician sınıfındaki test1 kontrol edelim:
# Sınıfta tanımlanan halinde gelir.

busra.test1()
````

````python
# mustafa objesini değer atayarak tanımlayalım:

mustafa = MusicianPlus("Mustafa")
````

````python
# MusicianPlus sınıfındaki değiştirilen test1 methodunu kontrol edelim:
# Değişim gerçekleşti.

mustafa.test1()
````

Gördüğümüz gibi; **Musician** sınıfında tanımlanan `test1` metodu, **MusicianPlus** sınıfı içinde aynı isimle tekrar tanımlandığında, bu işlem **`override`** olarak adlandırılır.

Böylece **MusicianPlus** nesnesi üzerinden çağrıldığında, üst sınıftaki değil, alt sınıftaki `test1` metodu çalışır.

## 2. Polymorphism (Çok Biçimlilik)

**Tanım:**

- Aynı metot isminin farklı sınıflarda farklı işlevlerde kullanılabilmesidir.

**Amaç:**

- Kodun esnekliğini ve genelliğini artırmak.

**Peki Ne İşe Yarar?**

- Kod daha esnek,

- Daha yeniden kullanılabilir (reusable),

- Daha ölçeklenebilir hale gelir.

Birden fazla sınıf içinde aynı isimde bir metod olabilir, ama bu metodlar her sınıfa özel farklı davranır.

**2 YaygınPolymorphism (Çok Biçimlilik) Türü:**

**1. Override (Üzerine yazma)**

Üst sınıftaki metodu, alt sınıfta aynı isimle yeniden tanımlarsın.

Bunu bir önceki konumuzda öğrendin. Bu da bir polymorphism türüdür!

````python
# Başka bir örnek ile bu yapıya tekrar bakalım:

class Animal():
    def __init__(self):                 #initialzier
        print("Hayvan sınıfı oluşturuldu.")

    # speak metodu tanımlayalım:
    def speak(self):
        print("Hayvan ses çıkarır.")

class Cat(Animal):     # override → polymorphism
    def speak(self):
      print("Miyav!")

class Bird(Animal):    # override → polymorphism
    def speak(self):
      print("Cik!")
````

````python
# hayvan isimli nesne ismi atayalım ve animal sınıfıyla eşleştirelim:

hayvan = Animal()


# speak metodu ile çalıştıralım:
hayvan.speak()
````

````python
#kedi isimli nesne ismi atayalım ve Cat sınıfıyla eşleştirelim:

kedi = Cat()

# speak metodu ile çalıştıralım:
kedi.speak()
````

````python
#kus isimli nesne ismi atayalım ve Bird sınıfıyla eşleştirelim:

kus = Bird()

# speak metodu ile çalıştıralım:
kus.speak()
````

**2. Aynı Metod İsmini Farklı Sınıflarda Kullanmak**
**(Interface Polymorphism)**

- Farklı sınıflar var.

- Bu sınıflar arasında kalıtım (inheritance) ilişkisi olması şart değil.

- Her sınıfta aynı isimde bir metod var.

- Bu metodun içerikleri farklı, ama çağırma şekli hep aynı.

Bu sayede, hangi sınıfın metodunu çağırdığımızı bilmeden aynı komutla çalıştırabiliriz.

````python
# Meyveler için sınıf oluşturalım:

class Banana():

  def __init__(self,name):      #initializer
    self.name = name

  def info(self):               #info methodu tanımladık
    return f"100 calories {self.name}"    #f-string kullanımı

class Apple():

  def __init__(self,name):       #initializer
    self.name = name

  def info(self):               #info methodu tanımladık
    return f"150 calories {self.name}"     #f-string kullanımı

````

````python
# Oluşturduğumuz Banana sınıfı için isim atamaları yapalım:

banana = Banana("Yellow Banana")
````

````python
# Oluşturduğumuz Apple sınıfı için isim atamaları yapalım:

apple = Apple("Red Apple")
````

````python
# Kontrol edelim:

banana.info()
````

````python
# Kontrol edelim:

apple.info()
````

Evet... Şuan kodlarımız doğru çalışıyor. Her bir kapsamda olan kod bloklarında doğru çalışıyor. Peki bunu daha anlaşılır hale nasıl getiririm, bir liste içinde farklı meyvelerin değerlerine ulaşabilir miyim ?

Deneyelim!

````python
# Meyve listesi oluşturalım:

fruitList= [banana,apple]
````

````python
# for döngüsü ile değerlere erişelim:

for fruit in fruitList:
  print(fruit.info())          # method çalıştırmak için () koymayı unutmayalım.
````

Gördüğümüz gibi, tek bir meyve listesi içindeki farklı meyvelerin ayrı ayrı kalori değerlerine ulaşabiliriz.


**Farkı:**

**Override polimorfizm** → Üst sınıftaki metodu, alt sınıfta yeniden tanımlarsın (inheritance ilişkisi vardır).

**Interface polimorfizm**→ Sınıflar arasında kalıtım şart değildir, sadece metod imzaları aynı olur.


| Özellik                 | **Meyve Örneği (Banana & Apple)**                                                                    | **Hayvan Örneği (Animal & Cat & Bird)**                                                  |
| ----------------------- | ---------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------- |
| **Polimorfizm Türü**   --> | **Interface Polymorphism** (Aynı metod ismini farklı sınıflarda kullanma)                            | **Override (Runtime Polymorphism)**                                                      |
| **Tanım**            -->   | Farklı, bağımsız sınıflar (`Banana`, `Apple`) aynı isimli metodu (`info`) içerir.                    | Alt sınıflar (`Cat`, `Bird`) üst sınıftaki (`Animal`) metodu (`speak`) yeniden tanımlar. |
| **Miras (Inheritance)** -->| Yok – `Banana` ve `Apple` birbirinden bağımsız sınıflar.                                             | Var – `Cat` ve `Bird` sınıfları `Animal`’dan miras alır.                                 |
| **Amaç**           -->     | Ortak isimli metodlarla farklı sınıfları aynı şekilde işleyebilmek.                                  | Üst sınıftan gelen davranışı (metodu) alt sınıfta özelleştirmek veya değiştirmek.        |
| **Nasıl Çalışır?**  -->    | Liste veya döngü ile tüm nesnelerde `info()` metodunu çağırabilirsin, her biri kendi sonucunu döner. | `Animal` tipinden beklenen `speak()` metodu, alt sınıfta kendi versiyonuyla çalışır.     |
| **Avantajı**      -->      | Miras gerekmeden ortak metod yapısı sağlar.                                                          | Kod tekrarını azaltır, üst sınıf davranışını özelleştirir.                               |
| **Örnek Kullanım**  -->    | `fruitList = [banana, apple]` → `for fruit in fruitList: print(fruit.info())`                        | `animals = [Cat(), Bird()]` → `for a in animals: a.speak()`                              |




## 3. Encapsulation (Kapsülleme)

**Tanım:**

Encapsulation, verileri (özellikleri) ve bu veriler üzerinde işlem yapan metodları tek bir yerde (sınıfta) toplama ve gerektiğinde dışarıya kapatma işlemidir.

**Amaç:**

- Dışarıdan doğrudan erişimi engelleyerek, veriyi korumak ve güvenliği sağlamak.

**Neden Gerekli?**

- Nesnenin iç verileri doğrudan değiştirilirse hatalar olabilir.

- Herkesin istediği gibi değiştirmesini engelleyip, kontrollü bir güncelleme sağlar.

- Kod güvenliğini artırır.



Kodlarımızı yazarken okunabilirliğe dikkat ettiğimiz kadar, değiştirilmemesi gerekli bilgilerin güvenliğine de dikkat etmemmiz gereklidir.

Örneğin,  bir proje üzerinde birden fazla kişi çalışabilir ve her biri kod üzerinde farklı geliştirmeler yapabilir. Bu durumda yanlışlıkla verilerin değiştirilmemesi veya başka işlemler yapılmaması için kodlara encapsulation işlemi uygulamak gerekir.

````python
# Örnek bir E-Ticaret sitesinde kullanıcı girişinde
# telefon bilgisi girmesiyle fiyat çıktısı elde edilmesi:


class Phone():

  def __init__(self, name, price):
    self.name = name
    self.price = price

  def info(self):
    print (f"{self.name} price is: {self.price}")     # f-string
````

````python
# isim ataması yapalım ve değer girelim:

iphone = Phone("İPhone 14" ,500)
````

````python
# Bilgi kontrolü:

iphone.info()
````

Burada kullanıcının girdiği telefon bilgisine karşı fiyat bilgisinin değiştirilmemesini isteriz. Bu durumda fiyat bilgisinin korunması gereklidir. Aksi halde, farklı bir fiyat ataması ile verimiz kolayca etkilenir.

````python
# price değerini değiştirelim:

iphone.price = 220
````

````python
# Değişikliği kontrol edelim:

iphone.info()
````

Görüldüğü gibi; price değeri 500 iken, 220 olarak değiştirildi. Bu durumda kullanıcı sorgulama yaptığında ekrana bu fiyat gözükür ve hataya sebep olur.

**Sorunun özeti:**

- Veri bütünlüğü bozuluyor → fiyatın gerçek değeriyle oynanabiliyor.

- Hatalara yol açıyor → yanlış veri ekrana yansırsa kullanıcıyı yanıltır.

- Güvenlik zafiyeti → e-ticaret sitesinde kötü niyetli kişiler fiyatı değiştirebilir.

**Fiyat(price) değişkenini nasıl güvenli hale getirebiliriz?**

`__price` tanımı ile.

````python
# price değiştirilemez hale getirme:

class Phone():

  def __init__(self, name, price):
    self.name = name
    self.__price = price     #self.price yerine self.__price yazılacak.

  def info(self):
    print (f"{self.name} price is: {self.__price}")     # f-string
````

````python
# isim ataması yapalım ve değer girelim:

iphone = Phone("İPhone 14" ,500)
````

````python
# Bilgi kontrolü:

iphone.info()
````

````python
# price değerini değiştirelim:

iphone.price = 220
````

````python
# Değişikliği kontrol edelim:

iphone.info()
````

Görüldüğü gibi, Fiyat(price) değişkenine atama yapılsa bile değiştirilemez hale geldi.

**Python’da Encapsulation**

Python’da **public**, **private** erişim belirleyiciler Java veya C# gibi katı kurallı değildir.

Ancak:

**Public**→ Her yerden erişilebilir: Örneğin; (self.deger)

**Private** → Çift alt çizgi ile başlar: Örneğin; (__deger) — Sınıf dışında doğrudan erişilemez.


Peki; **fiyat bilgisinde değişiklik yapmak istediğimizde nasıl yapabiliriz ?**

Başka bir method tanımı yaparak.



````python
class Phone():

  def __init__(self, name, price):
    self.name = name
    self.__price = price     #self.price yerine self.__price yazılacak.

  def info(self):
    print (f"{self.name} price is: {self.__price}")     # f-string

  def changePrice(self, price):
    self.__price = price
````

````python
# isim ataması yapalım ve değer girelim:

iphone = Phone("İPhone 14" ,500)
````

````python
# Bilgi kontrolü:

iphone.info()
````

````python
# Değişikliği deneyelim:

iphone.__price = 300
````

````python
# Kontrol edelim:
# Yine değer korundu.

iphone.info()
````

````python
# Değişikliği methodla deneyelim:
# Method için (değer ataması) şeklinde uygulamayı unutma.

iphone.changePrice(300)
````

````python
# Kontrol edelim:
# yeni fiyat atandı. 500 yerine 300 olarak yeni değer belirlendi.

iphone.info()
````

## 4. Abstraction (Soyutlama)

**Tanım:**

- **Gereksiz detayları gizleyip,** kullanıcının yalnızca gerekli olan kısımlara erişmesini sağlayan OOP prensibidir.

**Amaç:**

- Nesnelerin nasıl çalıştığına değil, ne yaptığına odaklanmaktır.

Yani; kullanıcı, arka plandaki karmaşık işleyişi bilmeden, nesnenin işlevlerini kullanabilir.


**Neden Gerekli?**

- Kullanıcıyı karmaşadan uzak tutar

- Gereksiz bilgiyi gizler

- Kodun **bakımını** ve **geliştirilmesini** kolaylaştırır

- **Arayüz (interface)** mantığıyla çalışır



**Python’da Abstraction**

Python’da abstraction genellikle abstract class ve abstract method kullanılarak yapılır.

Bunun için **abc (Abstract Base Class) modülü** kullanılır.

Bu modül Python ile yüklü halde gelir ve bunu içeri aktararak kullanılabilir hale getiririz.

````python
# Soyut sınıflar ve abstract method'lar kullanmak için gerekli modül:

from abc import ABC, abstractmethod

# Soyut (abstract) sınıf tanımlıyoruz.
class Car(ABC):          # ABC → Abstract Base Class

    @abstractmethod
    def max_speed(self):       # 'max_speed' ile soyut (abstract) metod tanımladık.
        pass
````

````python
# Kod çalıştırıldığında hata alırız.

# myCar = Car()
````

Hata neden aldık? Çünkü;

- **Abstraction (Soyutlama)** = Sadece ne yapılacağını söylersin, nasıl yapılacağını söylemezsin.

- Python’da `abstract class` (soyut sınıf) kullanarak bunu yaparız.

- Soyut sınıfta** abstract method** (içi boş metod) olur.

- Bu sınıftan **direkt nesne oluşturamazsın**, çünkü o metodun içi boş.

- O metodun içini alt sınıfta doldurmak zorundasın.

Yani; yeni bir sınıf oluşturup Car sınıfından **kalıtım(inheritance)** alarak oluşturabiliriz.

Tekrar deneyelim:

````python
# Soyut sınıflar ve abstract method'lar kullanmak için gerekli modül:

from abc import ABC, abstractmethod


# Soyut (abstract) sınıf tanımlıyoruz.
class Car(ABC):          # ABC → Abstract Base Class

    # Abstract method tanımı:
    # Burada 'max_speed' metodunun NE yapacağını söylüyoruz ama NASIL yapacağını söylemiyoruz.
    @abstractmethod
    def max_speed(self):
        pass     # Metodun içi boş, bu yüzden Car sınıfından nesne üretilemez.


# 'Tesla' sınıfı Car sınıfından kalıtım alıyor.
# Burada abstract method'un içini doldurmak ZORUNDAYIZ.

class Tesla(Car):         # kalıtım oluşturalım:
    def max_speed(self):
      print("200km")      # Artık 'nasıl yapılacağı' tanımlandı.
````

````python
# Tesla sınıfından nesne oluşturuyoruz:

tesla = Tesla()
````

````python
# Artık abstract method'un içi dolu olduğu için çağırabiliriz:

tesla.max_speed()
````

**Hata nedenini tekrar hatırlayalım:**

İlk denemede `Car` sınıfından doğrudan nesne oluşturmak istedik.
Ancak `max_speed()` metodunun içi boş (abstract) olduğu için Python dedi ki:

**"Bu metodun içi dolu değil, bu yüzden nesne oluşturamazsın. Önce alt sınıfta tanımla."**

Python’da bunu `pass` anahtar kelimesiyle belirtiyoruz.

O yüzden soyut `sınıf (Car)` → `somut sınıf (Tesla)` sıralamasını izledik.

## **Özet:**

**OOP (Nesne Yönelimli Programlama)**, kodu `özellikler (property)` ve `davranışlar (method)` barındıran nesneler üzerinden modelleyerek daha düzenli, güvenli ve tekrar kullanılabilir hale getiren bir yaklaşımdır.

**Dört temel prensibi vardır:**

- `Encapsulation (kapsülleme)` veriyi koruyup kontrollü erişim sağlar;

- `Inheritance (kalıtım)` bir sınıfın başka bir sınıfın özelliklerini miras alarak kod tekrarını azaltır;

- `Polymorphism (çok biçimlilik)` aynı metod adının farklı alt sınıflarda farklı davranmasını sağlar;

- `Abstraction (soyutlama)` ne yapılacağını tanımlar, nasıl yapılacağını alt sınıflara bırakır.


Bu prensipler birlikte kullanıldığında, büyük projelerde modüler, esnek ve bakımı kolay yazılımlar geliştirilir.

Örneğin; **Bir e-ticaret projesinde OOP’nin dört prensibi birlikte çalışabilir:**

1. **Encapsulation** ile kullanıcı şifreleri gizlenip yalnızca doğrulama metodu üzerinden erişilir,

2. **Inheritance** ile tüm ürün tipleri (Telefon, Bilgisayar, Kıyafet) ortak bir Product sınıfından türetilir,

3. **Polymorphism** sayesinde her ürün tipi kendi indirim hesabını farklı şekilde yapar,

4. **Abstraction** ile ise ödeme sistemi sadece “öde” metodunu tanımlar, detaylarını (kredi kartı, havale, kripto) alt sınıflar uygular.

Böylece proje hem güvenli hem de kolay genişletilebilir olur.
