## **Özel Methodlar**

**Özel Methodlar (Special/ Magic / Dunder Methods)**

Python’da `__init__`, `__str__`, `__repr__`, `__eq__` gibi çift alt çizgiyle başlayan methodlara özel method denir.

Sınıf davranışlarını özelleştirmek için kullanılırlar.

Böylece sınıf nesnelerini Python dilinin standart fonksiyonlarıyla uyumlu hale getiririz.

Örneğin;

`__init__(self, ...):`

- Nesne oluşturulurken çalışan başlatıcı (initializer) metottur.

- Nesne özelliklerini burada tanımlar ve başlatırsın.

- Yapıcı fonksiyon gibidir.

`__str__(self):`

- Nesne print() veya str() ile yazdırıldığında dönen okunabilir metin.

- İnsan dostu, kullanıcıya gösterilecek metni döndürür.

`__repr__(self):`

- Nesnenin geliştiriciye yönelik resmi temsili.

- repr() veya doğrudan etkileşimli ortamda nesne çağrıldığında gösterilir.

- Genellikle nesnenin oluşturulabilir hali olarak yazılır.

`__eq__(self, other):`

- Eşitlik kontrolü için kullanılır (== operatörü).

- İki nesnenin eşit olup olmadığını karşılaştırmak için özelleştirilir.

**Nasıl çalışıyor?**

- **try:** → Burada hata çıkma ihtimali olan kodu yazarız.

- **except ValueError:** → Sadece sayı çevirmede hata olursa burası çalışır.

- Program hata verse bile **kapanmaz**, kullanıcıya anlamlı bir mesaj gösterir.

Burada ValueError dışında başka hata da olabilir. Farklı hata çeşitleri aynı kodda olduğunda ya except yazıp diğer hataları belirtebilirz veya;



```
try:
  myAge = int(input("Yaşınızı giriniz: "))
  print(myAge *2)

except:
  print("Lütfen sayısal bir değer giriniz.")
```

sadece **except:** belirtip alınan her hatada mesaj yazmasını sağlayabiliriz.



**Peki, kullanıcı doğru girene kadar hata mesajı almasını nasıl sağlayabiliriz?**

- while döngüsü ile.

**Dikkat!**

Eğer sürekli çalışmasını istiyorsak ve doğru değer girildiğinde sonlanmasını istiyorsak `while & break` kullanımıyla kodu yazmalıyız. Aksi halde, sonsuz döngüye gireceği için kodlarımız kitlenebilir.

````python
# Kodu while ile deneyelim:

while True:            #while 1==1: ile aynı.

  try:
    myAge = int(input("Yaşınızı giriniz: "))
    print(myAge *2)
    break                     # belirtmezsek sonsuz döngü olur.

  except:
    print("Lütfen sayısal bir değer giriniz.")
````

**Nasıl çalışıyor?**

**1. while True: → Sonsuz döngü**

Kod sürekli çalışmaya devam eder, ta ki biz break ile durdurana kadar.
(`while True = while 1==1 mantığı`)

**2. try: bloğu**

Kullanıcıdan `input()` ile yaş bilgisi alınır.
`int()` ile tam sayıya çevrilmeye çalışılır:


```
myAge = int(input("Yaşınızı giriniz: "))
```


- Eğer kullanıcı `"29"` gibi sayısal bir değer girerse çevrim başarılı olur.

- Başarılı olursa `print(myAge * 2)` çalışır (yaşın iki katını yazdırır).

- Ardından `break` çalışır ve döngü tamamen sonlanır.

**3. except: bloğu**

Eğer kullanıcı `"Büşra"` veya `"29,"` gibi sayı olmayan ya da içinde harf/virgül bulunan bir değer girerse `int()` çevirimi **ValueError hatası** verir.
Bu hata `except:` bloğu tarafından yakalanır ve şu mesaj basılır:



```
print("Lütfen sayısal bir değer giriniz.")

```

Ardından döngü break olmadığı için tekrar başa döner ve kullanıcıdan yeniden giriş ister.


**4. Döngünün bitişi**

- Yanlış giriş → hata yakalanır → mesaj basılır → tekrar sorulur.

- Doğru giriş → çarpma yapılır → `break` ile döngü kırılır → program biter.


**Özet:**

Bu yapı, kullanıcı doğru veri girene kadar programın güvenli şekilde devam etmesini sağlar. Böylece uygulama **çökmek yerine** kullanıcıyı tekrar veri girmeye yönlendirir.

**try-except** dışında else ve **finally** keywordleri de kullanılır.

**finally** -> kodun her zaman çalışmasını sağlar.

Ancak, `try-except kullanımı` hata yönetimi açısından yeterli olacaktır.

````python
# Kodu tekrar yazalım:

while True:            #while 1==1: ile aynı.

  try:
    myAge = int(input("Yaşınızı giriniz: "))
    print(myAge *2)
    #break                    #veya

  except:
    print("Lütfen sayısal bir değer giriniz.")
  else:
    print("Tebrikler doğru veri girdiniz.")
    break                     # belirtmezsek sonsuz döngü olur
  finally:
    print("Sonlandı.")
````

**Bu kodda:**

- `try` → Kullanıcıdan yaş alınıp `int`'e dönüştürülüyor, 2 ile çarpılıp yazdırılıyor.

- `except` → Eğer sayı dışında bir şey girilirse (örn. `"Büşra"`,` "30?"`, `"30,5"`) hata oluşuyor ve `"Lütfen sayısal bir değer giriniz."` mesajı çıkıyor.

- `else` → `try` bloğu hatasız çalışırsa (örn. 29 girildiğinde) `"Tebrikler doğru veri girdiniz."` yazıyor ve break ile döngü bitiyor.

- `finally` → Hata olsa da olmasa da **her zaman** `"Sonlandı."` mesajını veriyor.

Kısacası, hatalı girişte **except + finally** çalışıyor, doğru girişte **else + finally**

**Python’da Dosyalarla Çalışmak**

Python’da dosya açmak için open() fonksiyonu kullanılır:

**Neden 2. okumada dosya boş gözüküyor?**

İkinci okumada dosyanın boş görünmesinin sebebi, **dosya okuma işleminin imleci (pointer) dosyanın sonuna kadar ilerlemiş olmasıdır.**

**Açıklama:**

- `myFile.read()` metodu, dosyanın başından sonuna kadar tüm içeriği okur ve dosya içindeki okuma imlecini **dosyanın sonuna** getirir.

- Bu yüzden, aynı dosya nesnesi üzerinde tekrar `read()` yaparsan, imleç zaten dosyanın sonunda olduğu için **okunacak yeni veri kalmaz,** sonuç boş string `('')` olur.


**Çözüm yolları:**

**1. Dosya imlecini başa almak:**

````python
myFile.seek(0)  # Dosyanın başına git
myFile.read()   # Tekrar oku, dosya içeriği gelir
````

**2. Dosyayı kapatıp yeniden açmak:**

````python
myFile.close()         #dosyayı kapat
myFile = open("myfile.txt")      #dosyayı tekrar aç
myFile.read()                    # dosyayı tekrar oku
````

**Özet:**

- `read() `metodu dosya içeriğini okurken imleci sona taşır.

- İkinci kez `read()` yapmak için imleci `seek(0)` ile başa almak gerekir.

- Aksi halde boş string döner çünkü okunacak veri kalmamıştır.

Ayrıca, tüm okuma işlemlerinden sonra `close() `ile dosyaları tekrar kapatmak gereklidir.

Peki, bu şekilde her seferinde dosyayı kapatıp tüm işlemleri tekrardan başlatmak veya imleci `seek(0)` kullanımı ile tekrar başa mı getireceğiz?

**`Bu manuel işlemlerin dışında başka bir yöntem yok mu?`**

- Evet, var. O da `with open()` kullanımı.

**Neden `with open()`?**

- Dosyayı açar, işlemleri yapar, dosyayı **otomatik olarak kapatır.**

- Dosya imlecini manuel yönetmene gerek kalmaz, çünkü genellikle işlem bittiğinde dosya kapandığı için sonraki açmada imleç baştan başlar.

- Kod daha temiz, okunabilir ve güvenlidir.

- Hata olsa bile dosyanın kapanması garanti edilir.

````python
# with ile kodu açalım:

with open("myfile.txt") as myFile:
    myContent = myFile.read()

# Dosya burada otomatik kapanır.
````

**ÖZET:**

**1. with open() ile dosya açma**

- Dosya açılır, blok içi kod çalışır, sonra dosya otomatik kapanır.

- Böylece close() yazmana gerek kalmaz, dosya her durumda güvenle kapanır.

**2."w" - Write modu**

- Dosyayı yazmak için açar.

- Dosyanın içeriğini tamamen siler ve üzerine yazar.

- Yeni veri eskisini siler ve yenisini yazar.

- Bu yüzden dikkatli kullanmak gerekir.

**3. "a" - Append modu**

- Dosyayı açar ve mevcut içeriğin sonuna ekleme yapar.

- Var olan veri korunur.

- Yeni veri, dosyanın sonuna eklenir.

**4. "r" - Read modu**

- Dosyayı sadece okumak için açar.

- Dosya üzerinde değişiklik yapılamaz.

- Dosyanın tüm içeriğini veya istediğin kısmını okuyabilirsin.

**Örnek tam akış:**



```
# Yazma modu: dosya önce temizlenir, sonra yazılır
with open("myfile.txt", "w") as f:
    f.write("test 4\n")

# Append modu: dosyanın sonuna ekleme yapılır
with open("myfile.txt", "a") as f:
    f.write("test 5\n")

# Okuma modu: dosyanın içeriği okunur
with open("myfile.txt", "r") as f:
    content = f.read()
    print(content)

```

````python
# Python 3.10 ve üzeri ile gelen özellik:


def process_value(value: int| str) -> str:        # int, str haricinde başka değerlerde verilebilir.
    if isinstance(value, int):                  # int sınıfından oluşan bir objemi ? kontrolü
        return f"Value is an integer: {value}"
    else:
        return f"Value is a string: {value}"
````

````python
# Kontrol edelim:

process_value(3)
````

````python
# Kontrol edelim:

process_value("merhaba")
````

** Açıklamalar:**

1. `value: int | str`

- Bu, value parametresinin int veya str tipinde olabileceğini söylüyor.

- `| işareti` `Union` anlamına geliyor (Python 3.10 ve üzeri).

- Önceki Python sürümlerinde bu şöyle yazılırdı:

```
from typing import Union
def process_value(value: Union[int, str]) -> str:

```
2. `-> str`

- Fonksiyonun string tipinde bir değer döndüreceğini söylüyor.

3. `isinstance(value, int)`

- Değişkenin tipini çalışma zamanında kontrol ediyoruz.

- Annotation sadece ipucu olduğundan, Python burada tip kontrolünü kendiliğinden yapmaz.


````python
# Python 3.9 ve üzeri ile gelen özellik:

from typing import List     # 'typing' modülünden List tipini içe aktarıyoruz.

def sum_list(numbers: List[int]) -> int:
    return sum(numbers)
````

````python
# Konu çalıştıralım:

numbers = [1, 2, 3, 4, 5]
total = sum_list(numbers)
total
````

**Açıklamalar:**

**1. from typing import List**

- List burada `typing modülü`nden geliyor ve **liste tipini**, içinde barındırdığı eleman tipleri ile birlikte belirtmemizi sağlıyor.

- `List[int]` → **"integer değerlerden oluşan bir liste"** anlamına gelir.

- Bu Python’a “numbers parametresi bir **integer listesi** olmalı” bilgisini verir.

**2. Fonksiyon imzası**`(def sum_list(numbers: List[int]) -> int)`
numbers: List[int] → Parametrenin tipi integer listesi olarak bekleniyor.

- `-> int` → Fonksiyonun dönüş tipinin **integer** olması bekleniyor.

- Bu sadece ipucudur. Python bu tipleri **runtime’da zorunlu kılmaz**, ama:

- Kod editörleri (VS Code, PyCharm vs.) tip uyumsuzluğu konusunda uyarı verir.

- `mypy` gibi tip denetleme araçları ile otomatik kontrol yapılabilir.

**Python 3.9 sonrası değişiklik:**

`typing.List` yerine` yerleşik list` ile tip belirtilebilir:




```
def sum_list(numbers: list[int]) -> int:
    return sum(numbers)

```
