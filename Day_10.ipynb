Objeleri oluştururken name, age ve gender verilmesini zorunlu hale getirmek için   `def __init__(self):` içinde ayrı bir parametre olarak tanımlamamız gerekli.

````python
# Kodumuzu  düzenleyelim:

class Person():
  name =""
  age = 0
  gender =""

  #method, initializer
  def __init__(self, nameInput, ageInput, genderInput):
    name = nameInput
    age = ageInput
    gender = genderInput
````

## Python’da Genel Sınıf Tanımı:

**Sınıf Tanımı:**

```
class SınıfAdı:
    def __init__(self, parametre1, parametre2, ...):
        self.özellik1 = parametre1
        self.özellik2 = parametre2
        # ...

    def metod_adi(self):
        # işlemler
        pass

```

**Açıklamalar:**


| Parça                     | Açıklama                                                          |
| ------------------------- | ----------------------------------------------------------------- |
| `class SınıfAdı:`         | Sınıf tanımlanır. İsim büyük harfle başlatılır (Python geleneği). |
| `def __init__(self, ...)` | Yapıcı metot (constructor). Nesne oluşturulunca otomatik çalışır. |
| `self.özellik = değer`    | Nesneye ait özellikler tanımlanır.                                |
| `def metod_adi(self):`    | Nesneye ait bir işlev/metot tanımlanır.                           |
| `self`                    | Nesnenin kendisini temsil eder (this gibi düşünebilirsin).        |


**Not:**

**Python 3 kullanıyorsan ve özel bir miras durumu yoksa şu yazımı kullan:**



```
class Araba:
    ...
```

**Eğer açık bir şekilde miras göstermek istersen:**



```
class Araba(object):  # veya başka bir sınıftan
    ...
```

Her iki yazım da aynı şekilde bir **sınıf** tanımlar.

--


**Peki neden iki farklı yazım var?**


| Yazım            | Açıklama                                                                                                                                                       |
| ---------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `class Araba:`   | Python 3'te **en sade ve tercih edilen** yazım biçimidir.                                                                                                      |
| `class Araba():` | Python 2 zamanlarında, sınıfın mutlaka bir üst sınıftan miras alması gerektiği düşünülüyordu. O zamanlar `()` içine `object` yazılırdı: `class Araba(object):` |

--

Python 3'te artık bu zorunluluk yok, çünkü **tüm sınıflar varsayılan olarak** `object` sınıfından miras alır.



## Sınıf Örnekleri:

````python
# Sınıf tanımlayalım:

class Dog():

  def __init__(self,age):     #initializer tanımı
    self.age = age
````

````python
# Parametre tanımlayalım ve değer verelim:

myDog = Dog(3)
````

````python
# Kontrol edelim:

myDog.age
````

İnsan parametresi tanımlayarak köpeğin yaşından insan yaşını nasıl tahmin edebiliriz?

1. Yöntem:
- humanAge metodu tanımı ile:

````python
# Sınıf tanımlayalım:

class Dog():

  year = 7               # özellik tanımlayalım.

  def __init__(self,age):     #initializer tanımı
    self.age = age

  def humanAge(self):         #humanAge metodu tanımlayalım
    return self.age * self.year     #self yerine Dog.year ile de ulaşılabilir.
````

````python
# Parametre tanımlayalım ve değer verelim:

myDog = Dog(3)
````

````python
# humanAge parametresini tanımlayalım:

myDog.humanAge()
````

2. Yöntem:
- Metod kullanmadan initializer içinde kullanıcıdan istemeden, sınıftan obje oluşturulduğunda çağrılabilir.

````python
# Sınıf tanımlayalım:

class Dog():

  year = 7               # özellik tanımlayalım.

  def __init__(self,age):     #initializer tanımı
    self.age = age
    self.dogHumanAge = age * self.year      #initializer içinde tanımlayarakta alabiliriz.

````

````python
# Parametre tanımlayalım ve değer verelim:

myDog = Dog(3)
````

````python
# doghumanAge parametresini tanımlayalım:

myDog.dogHumanAge
````

`print() `metodunu da initializer alanına yazdığımızda kodu her çalıştırdığımızda ilgili yazıyı ekrana yazdırabiliriz.

**Kısaca;** initializer bölümü sadece kullanıcıdan başta zorunlu değişken girdisi almak için kullanılmaz.

Sınıftaki oluşturulan her bir objeye ulaşmak içinde kullanılır.

````python
# Sınıf tanımlayalım:

class Dog():

  year = 7               # özellik tanımlayalım.

  def __init__(self,age):     #initializer tanımı
    self.age = age
    self.dogHumanAge = age * self.year      #initializer içinde tanımlayarakta alabiliriz.
    print("Dog metodu çalıştırıldı")
````

````python
# Parametre tanımlayalım ve değer verelim:

myDog = Dog(3)
````

Default değerleri doğrudan `def __init__(self,age):`fonksiyonu içinde de tanımlayabiliriz.

Örneğin:

````python
# Sınıf tanımlayalım:

class Dog():

  year = 7               # özellik tanımlayalım.

  def __init__(self,age = 5):     #initializer içinde default değer tanımı
    self.age = age
    self.dogHumanAge = age * self.year      #initializer içinde tanımlayarakta alabiliriz.
    print("Dog metodu çalıştırıldı")
````

````python
# Farklı bir parametre ismi tanımlayalım:
# parametreye değer vermedik, hata vermedi.

barley = Dog()
````

````python
# Değeri kontrol edelim:

barley.age
````

````python
# dogHumanAge değer kontrolü:

barley.dogHumanAge
````

````python
# Default değeri değiştirebiliriz:

barley = Dog(15)
````

````python
# Değeri yeni değerle kontrol edelim:

barley.age
````

````python
# dogHumanAge yeni değerle kontrolü:

barley.dogHumanAge
````

```

        self.age = age
        self.gender = gender

Kullanıcı Girişi:
ali = Person("Ali")  # Sadece isim girildi
print(ali.age)       # 0
print(ali.gender)    # Belirtilmedi

```
